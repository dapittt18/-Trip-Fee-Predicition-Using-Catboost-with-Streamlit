# 1. Import Libraries

```python
import numpy as np
import pandas as pd
import datetime
import time
import seaborn as sns
import matplotlib.pyplot as plt
from pymongo import MongoClient
%matplotlib inline
```

# 2. Connect to MongoDB Compass

```python
# Instantiate MongoDB client by localhost 27017 
client = MongoClient("mongodb://localhost:27017")

# Get database logee_trans
db = client.get_database("logee_trans")

# Retrieve collection deliveryDetails from database
deliv_details = db.get_collection("deliveryDetails")
```

# 3. Convert Data
Convert the retrieved .json format data into pandas dataframe.

```python
df_delivery = pd.DataFrame(list(deliv_details.find({})))
```

# 4. Exploratory Data Analysis (EDA)

## 1. Check Data

```python
print(df_delivery.info())
print(df_delivery.shape)
display(df_delivery.head(20))
```

## Data Manipulation

### 1. Subset Data 1

First step is to creat subset data that only consist of column argoCompanyName, truckCompanyName, vehicleGroupName, fulfillment, and createdAt.

```python
newdeliv = df_delivery[['cargoCompanyName','truckCompanyName','vehicleGroupName','fulfillment','createdAt']]
```

```python
# Check information
print(newdeliv.info())
```

```python
print(newdeliv.shape)
display(newdeliv.head())
```

### 2. Convert to pandas NA

Convert numpy nan format into pandas NA.

```python
newdeliv = newdeliv.fillna(pd.NA)
```

```python
newdeliv.head(500)
```

### 3. Subset Data 2

Filter only records that doesn't have nan (null) values in any column. Then, rearrange data in ascending sort based on column truckCompanyName.

```python
newdeliv_notna = newdeliv.dropna(thresh=5)
```

```python
newdeliv_notna.sort_values(by='truckCompanyName',ascending=True,inplace=True)
```

```python
newdeliv_notna.reset_index(inplace=True)
```

```python
newdeliv_notna.drop('index',axis=1,inplace=True)
```

```python
newdeliv_notna.head()
```

```python
newdeliv_notna.shape
```

### 4. Delete unnecessary space

```python
newdeliv_notna.loc[:,'cargoCompanyName'] = newdeliv_notna.cargoCompanyName.apply(lambda x : x.strip())
newdeliv_notna.loc[:,'truckCompanyName'] = newdeliv_notna.truckCompanyName.apply(lambda x : x.strip())
newdeliv_notna.loc[:,'vehicleGroupName'] = newdeliv_notna.vehicleGroupName.apply(lambda x : x.strip())
```

```python
newdeliv_notna.head(1500)
```

```python
newdeliv_notna.iloc[0,0]
```

### 5. Subset Data - 3

Create new subset data that only consist of column truckCompanyName,vehicleGroupName, and createdAt from data newdeliv_notna. 

Break data in column fulfillmnet into seperated columns. 

Concat the 2 new datas into a single new subset data.


```python
new_deliv = pd.concat([newdeliv_notna.drop(['cargoCompanyName','fulfillment'], axis=1), newdeliv_notna['fulfillment'].apply(pd.Series)], axis=1)
```

```python
print(new_deliv.shape)
display(new_deliv.head())
```

### 6. Subset Data 4

Create new subset data that only consist of column truckCompanyName,vehiclePoliceNum, and createdAt.

```python
new_deliv = new_deliv[['truckCompanyName','vehiclePoliceNum','createdAt']]
```

```python
new_deliv
```

### 7. Final Data

Get year, month, and date information from column createdAt.

```python
# Get only year, month, and date information
def dates(x):
    return x.split('T')[0]
```

```python
new_deliv.loc[:,'Date'] = new_deliv.createdAt.apply(lambda x : dates(x))
```
