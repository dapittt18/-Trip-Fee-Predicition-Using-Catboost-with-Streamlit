# 2. Connect to MongoDB Compass

```python
# Instantiate MongoDB client by localhost 27017 
client = MongoClient("mongodb://localhost:27017")

# Get database logee_trans
db = client.get_database("logee_trans")

# Retrieve collection companies from database
company = db.get_collection("companies")
```

# 3. Convert Data

Convert the retrieved .json format data into pandas dataframe.

```python
companies = pd.DataFrame(list(company.find({})))
```

```python
df_companies = companies.copy()
display(df_companies.head())
print(df_companies.shape)
```

# 4. Import Dummy Data

This data contains list of cargo and truck company names that needs to be excluded while doing EDA process later.


```python
df_dummy = pd.read_excel('Data Dummy Perusahaan.xlsx')
df_dummy['CompanyID'] = df_dummy['CompanyID'].str.strip()
display(df_dummy)
print(df_dummy.shape)
```

# 4. Create list of companyId

Create a new list object that contains companyId's in dummy data.

```python
listdummy = list(df_dummy.CompanyID)
print(listdummy)
print("\n=====\n",len(listdummy))
```

# Data Manipulation

## 1. Subset Data 1

Subset data of companies in data dummy.

```python
df_companiesck = df_companies.loc[df_companies['companyId'].isin(listdummy)]
display(df_companiesck)
print(df_companiesck.shape)
```

## 2. Subset Data 2

Subset data of companies not in data dummy.

```python
df_companiescc = df_companies.loc[~df_companies['companyId'].isin(listdummy)]
display(df_companiescc)
print(df_companiescc.shape)
```

## 3. Subset Data 3

From subset data 2, pick only company that act as a cargoOwner or in other terms have cargoOwner partnership type.

```python
df_co = df_companiescc.loc[df_companies['partnershipType'] == 'cargoOwner']
display(df_co)
print(df_co.shape)
```

## 4. Subset Data 4

From subset data 3, pick only companies that don't have id partner/ invalid id partner value.

```python
df_cona = df_co.loc[df_co['partnerIds'].str.len() == 0]
display(df_cona)
print(df_cona.shape)
```

## 5. Subset Data 5

From subset data 3, pick only companies that have id partner/ valid id partner value.

```python
df_conotna = df_co.loc[df_co['partnerIds'].str.len() != 0]
display(df_conotna)
print(df_conotna.shape)
```

## 6. Extract partnerIds column

The partnerIds column itself has data in the form of list, where each list contains data such as company id, company name, and partner id of the company. Therefore, we need to break down these three data into 3 separate columns.

```python
df_cocc = df_conotna.explode('partnerIds')
df_cocc = df_cocc[['companyId', 'name', 'partnerIds']][:]
display(df_cocc)
```

## 6. Subset Data 6

From subset data 2, pick only companies data that act as a truck owner (has truck owner as the partnership type). From that data, subset only column companyId and name.

```python
df_to = df_companiescc.loc[df_companies['partnershipType'] == 'truckOwner', ['companyId','name']]
display(df_to)
print(df_to.shape)
```

## 7. Subset Data 7

From subset data 6, pick only data of cargo companies which the companyId doesn't exist on column partnerIds from subset data df_cocc.

```python
df_tonotrel = df_to.loc[~df_to['companyId'].isin(df_cocc.partnerIds)]
display(df_tonotrel)
print(df_tonotrel.shape)
```

## 8. Rearrange Data

Sort the data in ascending order based on company name.

```python
df_tonotrel.sort_values(by='name',ascending=True,inplace=True)
```

```python
df_tonotrel.reset_index(inplace=True)
```

```python
df_tonotrel.drop('index',axis=1,inplace=True)
```

```python
df_tonotrel
```

## 14. Concat Data

This is the first step to create the final data product. First, set column partnerIds of data df_cocc into the data index, then concat it with data df_to. 

Then, add suffix _co for cargo companies column name and _to for truck companies column name.

```python
df_mapping = df_cocc.set_index('partnerIds').join(df_to.set_index('companyId'), lsuffix='_co', rsuffix='_to')
```

```python
df_mapping.reset_index(inplace=True)
```

```python
df_mapping
```

```python
df_mapping = df_mapping[['companyId','name_co','partnerId','name_to']]
```

```python
df_mapping = df_mapping.sort_values(by='name_co',ascending=True).reset_index().drop(['index'],axis=1)
```

```python
print(df_mapping.shape)
display(df_mapping)
```

## Drop duplicates

Drop duplicates based on subset column partnerId and companyId.

```python
df_mapping = df_mapping.drop_duplicates(subset = ["partnerId","companyId"])
df_mapping
```

```python
df_mapping = df_mapping.reset_index().drop('index',axis = 1)
df_mapping
```

## Final Data

Dataframe df_mapping is the data that stakeholders want. To create the final data product,group the data based on the companyId, name_co, partnerId, and name_to columns.

Then sort the data by column name_co in ascending order and save the result in the df_map_group dataframe.

```python
df_map_group = df_mapping.groupby(['companyId','name_co','partnerId',
                        'name_to']).count().sort_values(by='name_co',ascending=True)
```

```python
df_map_group
```

Berdasarkan data di atas, dapat dilihat data pemetaan antara perusahaan cargo dan perusahaan truk beserta id dari setiap perusahaan terkait. Hasil menunjukkan bahwa beberapa perusahaan cargo memiliki mitra dengan lebih dari 1 perusahaan truk. Kemudian ada juga perusahaan cargo yang bermitra dengan 1 perusahaan truk saja.

Based on the data above, the results show that some cargo companies have partners with more than 1 trucking company. Then there are also cargo companies that paired with only one trucking company.
