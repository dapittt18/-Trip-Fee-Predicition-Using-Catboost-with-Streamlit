# Data Manipulation

## 1. Filter Dummy

Disini kita akan menggunakan data dummy untuk mem-filter perusahaan pada data df_delivery.

```python
# Delete trailing space
dummy[['CompanyID','Nama Perusahaan']] = dummy[['CompanyID','Nama Perusahaan']].applymap(lambda x : x.strip())
# Upper companies name
dummy.loc[:,'Nama Perusahaan'] = dummy['Nama Perusahaan'].apply(lambda x : x.upper())
```

```python
dummy.head(10)
```

## 2. Create List

Create a new list that contains companies name from dummy data.

```python
cargodummy = list(dummy['Nama Perusahaan'].unique())
new_cargodummy = []

for item in cargodummy:
    new_cargodummy.append(item)
display(new_cargodummy)
idummy = list(dummy['CompanyID'].unique())
display(idummy)
```

## 3. Data Transformation 1

In this first data transformation, we will 

1. Replace null values in column cargoCompanyName and truckCompanyName into 'Other'.

2. Delete space at the start and end of every cargo and truck company name.

3. UPPER all company names.

```python
df_delivery['cargoCompanyName'].replace(np.nan,'Other',inplace=True)
df_delivery['truckCompanyName'].replace(np.nan,'Other',inplace=True)
df_delivery[['cargoCompanyName','truckCompanyName']] = df_delivery[['cargoCompanyName',
                                            'truckCompanyName']].applymap(lambda x : x.strip())
df_delivery[['cargoCompanyName','truckCompanyName']] = df_delivery[['cargoCompanyName',
                                            'truckCompanyName']].applymap(lambda x : x.upper())
```

```python
df_delivery[['cargoCompanyName','truckCompanyName']] 
```

## 4. Subset Data 1

Subset data of cargo and truck companies not in data dummy.

```python
df_delivery = df_delivery[~(df_delivery['cargoCompanyId'].isin(idummy))]
df_delivery = df_delivery[~(df_delivery['truckCompanyId'].isin(idummy))]
```

```python
df_delivery.shape
```

## 5. Transformasi Data 2

Because cargoCompanyName and truckCompanyName column have been filtered out from dummy, the next step is to revert 'Other' value back into its origin nan values.

```python
df_delivery['cargoCompanyName'].replace('Other', np.nan ,inplace=True)
```

```python
df_delivery['truckCompanyName'].replace('Other', np.nan ,inplace=True)
```

## 9. Delete Missing Values

In this task, we need to find the number of transactions made by each user of each cargo company. Therefore, the columns that are important to note are:

1. cargoCompanyId
2. cargoCompanyName
3. createdBy
4. totalAmount
5. tripFeeAmount
6. taxAmount

In this step, we will also check for null values in each of these columns.

```python
df_delivery[['cargoCompanyId','cargoCompanyName','createdBy','totalAmount','tripFeeAmount','taxAmount']].isna().sum()
```

There are 9 empty rows of data in the cargoCompanyId, createdBy, totalAmount, tripFeeAmount, and taxAmount columns. Therefore, the next step is to retrieve records where there are no empty values ​​in all of the columns.



```python
delivs_notna = df_delivery.dropna(thresh=6,subset= ['cargoCompanyId','cargoCompanyName','createdBy',
                    'totalAmount','tripFeeAmount','taxAmount'])
delivs_notna.shape
```

## 10. Subset Data 2

Create new subset data that only consist of column cargoCompanyId, cargoCompanyName, createdBy, and totalAmount. 

```python
delivs_notna = delivs_notna[['cargoCompanyId','cargoCompanyName','createdBy','totalAmount']]
print(delivs_notna.shape)
display(delivs_notna)
```

```python
delivs_notna.reset_index(inplace=True)
```

```python
delivs_notna.drop('index',axis=1,inplace=True)
```

```python
delivs_notna
```

## 11. Import User Data

User data is required to get user name for every order.

```python
client = MongoClient("mongodb://localhost:27017")
db = client.get_database("logee_trans")
user = db.get_collection("user")
```

```python
df_user = pd.DataFrame(list(user.find({})))
print(df_user.shape)
display(df_user)
```

## 12. Create List

Create a new list that contains pairs of user name and its corresponding user id.

```python
users = df_user[['userId','name']]
pengguna = []
for x,y in zip(list(users.userId.values), list(users.name.values)):
    pengguna.append((x,y))
pengguna
```

```python
len(pengguna)
```

## 13. Data Transformation 3

```python
# Function to get name from every id in createdBy column
def name(x):
    for ids,name in pengguna:
        if x == ids:
            return name
        else :
            pass
delivs_notna.loc[:,'name'] = delivs_notna.createdBy.apply(lambda x : name(x))
```

```python
delivs_notna.head(1000)
```

Values in totalAmount column are still in float. Convert it into integer.

```python
delivs_notna['totalAmount'] = delivs_notna['totalAmount'].astype(np.int64)
```

```python
delivs_notna
```

```python
delivs_notna.sort_values(by='cargoCompanyName',ascending=True,inplace=True)
```

```python
delivs_notna
```

Delete unnecessary space in cargoCompanName column.

```python
delivs_notna.loc[:,'cargoCompanyName'] = delivs_notna.cargoCompanyName.apply(lambda x : x.strip())
delivs_notna.sort_values(by='cargoCompanyName',ascending=True,inplace=True)
delivs_notna
```

## 14. Data Grouping

In this step, the data will be grouped based on cargoCompanyName, name, and createdBy columns. After that, we do sum and count on totalAmount column. This is a mandatory step because data grouping requires aggregation process on data.

```python
delivery_final = delivs_notna.groupby(['cargoCompanyName','name','createdBy']).agg({'totalAmount': ['sum','count']})
delivery_final
```

```python
delivery_final.reset_index(inplace=True)
delivery_final
```

```python
delivery_final.columns = ['cargoCompanyName','name','Id','Total_transaction','Frequency']
```

```python
delivery_final
```

Now, grouped data again based on cargoCompanyName, name, Id, Total_transaction, and Frequency columns.

```python
delivery_final_group = delivery_final.groupby(['cargoCompanyName','name','Id','Total_transaction','Frequency']).count()
delivery_final_group
```
